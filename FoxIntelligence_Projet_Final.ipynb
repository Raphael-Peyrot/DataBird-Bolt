# Nettoyage et mise en forme de la donnée

Le fichier "transactions" ou dans le cadre du coding le dataframe "df3" comprend 95% des informations avec lesquelles nous allons travailler. Les informations utiles dans les autres fichiers "customers" et "products" sont directement exploitables (count_customers et product_name) et n'ont pas besoin d'être nettoyées.

```python
# On change le nom des colonnes pour simplifier la lecture.
df3 = df3.rename(columns={'synthetic_gender': 'genre','synthetic_age_generation' : 'generation', 'merchant_name' : 'compagnie_livraison'})
df3 = df3.rename(columns={'order_date' : 'date_commande', 'order_time' : 'heure_commande', 'order_total_paid' : 'prix_commande' })
df3 = df3.rename(columns={'order_currency' : 'devise', 'payment_method_name' : 'moyen_de_paiement', 'order_total_promo' : 'montant_promo' })
df3 = df3.rename(columns={'seller_name' : 'nom_restaurant' , 'seller_category_name' : 'type_de_nourriture', 'order_item_delivery_address_region' : 'region_livraison'})
df3 = df3.rename(columns={'order_item_delivery_address_subregion' : 'departement_livraison', 'order_item_delivery_address_city' : 'ville_livraison'})
```

```python
# On filtre les régions non françaises:
df3 = df3[(df3['region_livraison'] != 'canarias' ) & (df3['region_livraison'] != 'cataluna' ) & (df3['region_livraison'] != 'greater london' ) ]
```

```python
# On crée une nouvelle colonne pour déterminer la valeur totale des commandes avec une promotion pour la société de service de livraison.
df3['total_avec_promo'] = df3['montant_promo'] + df3['prix_commande']
```

```python
# On supprime dans la colonne "total_avec_promo" les commandes dont le prix est inférieur à 1€
# (il s'agit uniquement des commandes de JustEat) car ces données sont erronées.
df3 = df3[df3['total_avec_promo'] > 1]
```

```python
# On filtre les NaNs dans la colonne "prix_commande".
df3 = (df3[df3['prix_commande'].notna()])
```

```python
# On change le type de date commande et heure commande en format datetime afin de pouvoir les manipuler plus tard.
df3['date_commande'] = pd.to_datetime(df3['date_commande'])
df3['heure_commande'] = pd.to_datetime(df3['heure_commande'])
```

```python
# Création des colonnes temporelles:
df3['annee_mois'] = df3['date_commande'].dt.to_period('M')
df3['jour_semaine'] = df3['date_commande'].dt.day_name()
df3['heure_commande'] = df3['heure_commande'].dt.time
df3['annee'] = df3['date_commande'].dt.year
df3['mois'] = df3['date_commande'].dt.month
```

```python
# On change le type de la colonne 'annee' en datetime afin de pouvoir la manipuler plus tard.
# df3['annee'] = pd.to_datetime(df3['annee'])
# df3['annee'] = df3['annee'].dt.to_period('Y')
```

```python
# On drop les colonnes inutiles de df3:
df3.drop(columns=['row_nb','is_loyalty_deliveroo', 'devise'], axis=1, inplace=True)
# 'row_nb' est inutile, 'is_loyalty_deliveroo' comprend énormément de NaNs et ne nous intéresse pas, 'devise' a toujours la même valeur (€)
```

```python
# Dans la colonne synthetic_gender on remplace 1.0 par 'Homme' et 2.0 par 'Femme'et NaN par 'Non renseigné'.
df3['genre'] = df3['genre'].replace({1.0: 'Homme', 2.0: 'Femme'})
df3['genre'] = df3['genre'].fillna('Non renseigné')
```

```python
# On remplace les NaNs par "Non renseigné" ou "Autre" pour les gérer et garder le reste de l'information dont on dispose.
df3['region_livraison'] = df3['region_livraison'].fillna('Non renseigné')
df3['ville_livraison'] = df3['ville_livraison'].fillna('Non renseigné')
df3['departement_livraison'] = df3['departement_livraison'].fillna('Non renseigné')
df3['generation'] = df3['generation'].fillna('Non renseigné')
df3['type_de_nourriture'] = df3['type_de_nourriture'].fillna('Autre')
```

```python
# On renomme et groupe les types de nourriture pour rendre les informations plus lisibles.
rename =  {'north american - chicken' : 'Fast food Américain', 'north american - sandwich':'Fast food Américain',
           'north american - bbq':'Fast food Américain' , 'north american - bagel' : 'Fast food Américain',
           'north american - others' : 'Fast food Américain', 'north american - burger' : 'Burger',
           'north american - hawaiian' : 'Hawaïen', 'european - italian - pizza' : 'Pizza',
           'european - italian - pasta' : 'Italien', 'european - italian - others' : 'Italien',
           'latin american - mexican' : 'Mexicain' ,'latin american - others' : 'Sud Américain',
           'latin american - peruvian' : 'Sud Américain', 'asian - japanese - sushi' : 'Japonais',
           'asian - japanese - others' : 'Japonais', 'european - french' : 'Français',
           'asian - thai' : 'Thaïlandais', 'asian - indian' : 'Indien' , 'middle eastern - kebab' : 'Kebab',
           'middle eastern - others' : 'Kebab', 'asian - vietnamese' : 'Vietnamien', 'asian - chinese': 'Chinois',
           'middle eastern - lebanese' : 'Libanais', 'european - salad' : 'Salade', 'asian - korean' : 'Coréen',
           'european - spanish' : 'Espagnol', 'european - british' : 'Britannique', 'european - others': 'Européen',
           'asian - others' : 'Divers Asiatique', 'others' : 'Autre', 'african' : 'Africain' }
df3['type_de_nourriture'] = df3['type_de_nourriture'].replace(rename)
```

```python
# On renomme et groupe le nom des générations pour rendre les informations plus lisibles.
rename =  { 'Gen. Y - [1980-1994]': '25-40 ans', 'Gen. Z - [1995-2010]':'18-25 ans',
           'Gen. X - [1960-1979]':'40- 60 ans' , 'Boomer - [1940-1959]' : 'Seniors',
           'Pre boomer - [1920-1939]' : 'Seniors' }
df3['generation'] = df3['generation'].replace(rename)
```

## **Analyse marché**

```python
# Parts de marchés des acteurs majeurs de la livraison:
# Calcul de la somme des prix de livraison pour chaque compagnie de livraison:
ca_compagnie = df3.groupby('compagnie_livraison')['prix_commande'].sum()
# Calcul du total des prix de commande:
total_prix_commande = ca_compagnie.sum()
# Calcul des parts de marché
part_de_marche = round((ca_compagnie / total_prix_commande) * 100, 2)
part_de_marche
```
